```python
import pandas as pd
```

```python
df = pd.read_parquet('yellow_tripdata_2023-01.parquet')
```

```python
df
```

```python
print(pd.io.sql.get_schema(df,name='yellow_taxi_data'))
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
df.to_sql(name='yellow_taxi_data',con=engine,if_exists='replace')
```

```python
%time df.head(0).to_sql(name='yellow_taxi_data',con=engine,if_exists='replace')
```

```python
from time import time
import pyarrow.parquet as pq
parquet_file = pq.ParquetFile('yellow_tripdata_2023-01.parquet')
for i in parquet_file.iter_batches():
    t_start = time()
    
    pd_i = i.to_pandas()
    pd_i.to_sql(name='yellow_taxi_data',con=engine,if_exists='append')
    t_end = time()
    print("Batch sent in : ",t_end - t_start)
```
